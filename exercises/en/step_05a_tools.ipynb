# Step 5a — Tools

🇬🇧 **English** (this notebook)

Give an agent a tool to call at runtime — live web search, in this case. The agent decides *when* to call the tool and *what query* to use; this is what turns a frozen-knowledge model into one that can ground its answer in current information. Standalone: the tool is attached directly to the `Agent`, no `Crew` needed.

## The exercise

`crewai_tools` ships `SerperDevTool` for web search out of the box — pass it directly to `tools=[...]` on the `Agent`. Needs `SERPER_API_KEY` in your `.env` (free key at [serper.dev](https://serper.dev)).

```python
import os

from dotenv import load_dotenv
from crewai import Agent
from crewai_tools import SerperDevTool

load_dotenv()

role      = "TODO: who is this agent? (something that plausibly needs current information)"
goal      = "TODO: what is this agent trying to achieve?"
backstory = "TODO: what expertise does this agent bring?"

agent = Agent(
    role=role,
    goal=goal,
    backstory=backstory,
    tools=[SerperDevTool()],
    verbose=True,
)

# ── A question that needs live information, not training-data knowledge ──────
question = "TODO: ask something about your topic that requires current/live information"

result = agent.kickoff(question)
print(result.raw)
```
