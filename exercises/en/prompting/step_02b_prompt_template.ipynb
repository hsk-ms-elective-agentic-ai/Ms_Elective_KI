- **`system`**: tells the model who it is and how to behave (persona, rules, format)

The system message is invisible to the end user but shapes every response. Try removing individual components to see what each one actually does.

## Background

A prompt is a program. The strategies you apply — providing examples, structuring inputs into named components, splitting a task into sequential calls, asking for explicit reasoning — are the code. The model is the runtime.

> Liu, P., Yuan, W., Fu, J., Jiang, Z., Hayashi, H., & Neubig, G. (2023). *Pre-train, Prompt, and Predict: A Systematic Survey of Prompting Methods in Natural Language Processing*. ACM Computing Surveys, 55(9), 1–35. [arXiv:2107.13586](https://arxiv.org/abs/2107.13586)

## Setup

Run this once before the exercise cell below.

```python
import os

from dotenv import load_dotenv
from crewai import LLM
from IPython.display import Markdown, display

load_dotenv()

llm = LLM(model=os.getenv("MODEL", "gemini/gemini-2.5-flash"))
```

## The exercise

One API call, no examples, no reasoning instruction — but now the message is split across two *roles*:

- **`system`**: background instructions the end user never sees — `persona`, `instruction`, `context`, `data_format`, `audience`, `tone`. This tells the model who it is and how it should behave for every response in the conversation.
- **`user`**: the actual question — just the topic you want answered.

This is the standard structure of the OpenAI Chat Completions API (and any model that follows it). The `system` role is a first-class concept in the protocol, not just a formatting convention.

Below are two fully worked examples — same question, two different personas — instead of a blank template to fill in.

The persona component alone can shift the whole response: "You are a patient high school teacher" versus "You are a senior machine learning engineer reviewing a pull request" will produce very different tone, vocabulary, and depth for the exact same question and topic.

```python
# ── Example 1: teacher persona ────────────────────────────────────────────
persona_1     = "You are a patient high school teacher.\n"
instruction_1 = "Explain the topic clearly, using a simple, relatable analogy.\n"
context_1     = "Your students have no prior background in the subject.\n"
data_format_1 = "Keep it to one short paragraph.\n"
audience_1    = "The reader is a curious 15-year-old.\n"
tone_1        = "Be warm and encouraging.\n"

system_message_1 = persona_1 + instruction_1 + context_1 + data_format_1 + audience_1 + tone_1

# ── Example 2: engineer persona ───────────────────────────────────────────
persona_2     = "You are a senior machine learning engineer.\n"
instruction_2 = "Explain the topic precisely, focusing on the technical and compliance implications.\n"
context_2     = "Your audience already works professionally with AI systems.\n"
data_format_2 = "Keep it to one short paragraph.\n"
audience_2    = "The reader is a fellow engineer assessing what needs to change in the product.\n"
tone_2        = "Be concise and matter-of-fact.\n"

system_message_2 = persona_2 + instruction_2 + context_2 + data_format_2 + audience_2 + tone_2

# ── Same topic and question, both times — only the system message changes ────
user_message = "Explain the EU AI Act in simple terms in a short paragraph."

output_1 = llm.call(
    messages=[
        {"role": "system", "content": system_message_1},
        {"role": "user", "content": user_message},
    ]
)
output_2 = llm.call(
    messages=[
        {"role": "system", "content": system_message_2},
        {"role": "user", "content": user_message},
    ]
)

display(Markdown(f"## Example 1 — Teacher persona\n\n{output_1}"))
display(Markdown(f"## Example 2 — Engineer persona\n\n{output_2}"))
```
