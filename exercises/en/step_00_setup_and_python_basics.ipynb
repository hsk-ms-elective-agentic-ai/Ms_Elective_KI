## 2. The `uv` Package Manager

This project uses [`uv`](https://docs.astral.sh/uv/) — a fast, modern Python package manager — instead of plain `pip` + `venv`. It manages the virtual environment and every dependency for you from a single `pyproject.toml` file.

### Installing uv

```bash
# macOS / Linux
curl -LsSf https://astral.sh/uv/install.sh | sh

# Windows (PowerShell)
powershell -c "irm https://astral.sh/uv/install.ps1 | iex"
```

Verify it installed:

```bash
uv --version
```

### Setting up this project

From your team repo's root:

```bash

```

This creates a `.venv/` and installs every dependency listed in `pyproject.toml` — `crewai` (agents, tasks, and the `LLM` class you'll use directly in steps 1-2), `jupyter`/`ipykernel` (these notebooks), `python-dotenv` (`.env` loading), and `google-generativeai`. Nothing to install one by one.

Then copy `.env.example` to `.env` and fill in your API keys (see the main README's [Technical Setup](../../README.md#4-technical-setup) section).

Run any command through `uv` without activating anything manually:

## 4. Python Basics

A fast refresher on the language itself, independent of any library. If you're new to Python, work through every cell; if not, skim for anything unfamiliar.

### Values: `int`, `float`, `str`

````python
i = 5
f = 5.0
s = "data science"

print(i, type(i))
print(f, type(f))
print(s, type(s))
````

Python works fine as a calculator, right out of the box:

````python
print(3 + 5)
print(2 / 3)
print((3 + 5) * 2 / 3)
print(3 % 2)  # modulo — the remainder of a division
````

Strings can be concatenated with `+` (no automatic space) or repeated with `*`:

````python
a = "data science"
b = "and machine learning"

print(a + b)      # no space
print(a + " " + b)
print(a * 3)
````

### Lists

An ordered, mutable sequence — can mix types, indexed from `0`, and sliceable with `[start:stop]`.

````python
items = ["a", "b", 5, "c", 6, 7.6]

print(items[0])        # first element
print(items[1:4])      # second through fourth

items[0] = "two"       # modify in place
print(items)

items.extend([1, 2, 3])  # append multiple items
print(items)
````

### Dictionaries

An unordered collection of `key: value` pairs — fast lookup by key instead of position.

````python
ages = {"Ana": 24, "John": 20, "Li": 27}

print(ages["Li"])      # look up by key

ages["Karsten"] = 27   # add a new key
print(ages)

empty = {}              # start empty, populate later
empty["key1"] = 1
print(empty)
````

### Getting help

`help(...)` prints a function or type's documentation right in the notebook. In VS Code/Cursor, `Shift+Tab` while your cursor is inside a function call does the same thing as a quick popup.

````python
help(str.upper)
````

### Control flow: `for` loops

Repeats an action once per element of a sequence. The loop variable (`i` below, by convention — name it anything meaningful) is available inside the loop body.

````python
for i in range(5):
    print("step", i)
````

### Control flow: `if` / `elif` / `else`

Runs a branch only when its condition is true — combine with a `for` loop to filter a sequence.

````python
for i in range(10):
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
````

### Functions

A function encapsulates a task so you can call it (with different arguments) instead of copy-pasting the code every time. Document what it expects and returns with a docstring (the string right after `def ...:`) — `help()` and `Shift+Tab` both read this back to you.

````python
def greet_n_times(name: str, n: int) -> None:
    """Print a greeting for `name`, `n` times.

    Arguments
    ---------
    name : str
        Who to greet.
    n : int
        How many times to print the greeting.
    """
    for _ in range(n):
        print(f"Hello, {name}!")


greet_n_times("research_crew", 3)
````

````python
help(greet_n_times)
````
